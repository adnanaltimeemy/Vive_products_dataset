```python
# zx_cola_dataset_example.py
import pandas as pd
import numpy as np

# ---------------------------
# Configuration: Regions and years
# ---------------------------
regions = ["United Kingdom", "United States"]
years = [2024, 2023, 2022, 2021]  # priority order shown, but dataset covers all

# ---------------------------
# 1) Sales and profit margins table
# ---------------------------
# Example columns:
#   year, region, currency, total_sales, profit_margin_pct, units_sold, avg_price_per_unit
rows = []
np.random.seed(42)  # deterministic example values; remove or change seed for variation

for region in regions:
    for year in years:
        # Example synthetic values — replace with real data
        # Suppose total sales in USD or GBP; choose currency by region
        currency = "GBP" if region == "United Kingdom" else "USD"
        # Synthetic sales: random around some baseline
        baseline = 5_000_000 if region == "United Kingdom" else 12_000_000
        year_offset = (year - 2021)  # 0..3
        sales = baseline * (1 + 0.05 * year_offset) * np.random.uniform(0.9, 1.1)
        # Profit margin synthetic: between 8% and 20%
        profit_margin = np.random.uniform(8.0, 20.0)
        # Units sold synthetic: assume avg price, compute units
        avg_price = 0.50 if region == "United Kingdom" else 0.75
        units_sold = sales / avg_price
        
        rows.append({
            "year": year,
            "region": region,
            "currency": currency,
            "total_sales": round(sales, 2),
            "profit_margin_pct": round(profit_margin, 2),
            "units_sold": int(round(units_sold)),
            "avg_price_per_unit": round(avg_price, 2),
        })

df_sales = pd.DataFrame(rows)
df_sales = df_sales.sort_values(["region", "year"], ascending=[True, False])

# Save to CSV
df_sales.to_csv("zx_cola_sales_profit.csv", index=False)

# ---------------------------
# 2) Demographics table
# ---------------------------
# Options per region × year: breakdown by Age Group, Gender, Income Bracket, etc.
# Each row: year, region, demographic_category, demographic_value, share_pct, or count
demographic_categories = {
    "Age Group": ["18-24", "25-34", "35-44", "45-54", "55+"],
    "Gender": ["Male", "Female", "Other"],
    "Income Bracket": ["Low", "Medium", "High"],
    # add more categories as needed
}

dem_rows = []

for region in regions:
    for year in years:
        # Example re-use of units sold to estimate counts; optional
        row_sales = df_sales[
            (df_sales.region == region) & (df_sales.year == year)
        ].iloc[0]
        units = row_sales["units_sold"]
        
        for cat, values in demographic_categories.items():
            # Random share assignment that sums to 100 for each category × region × year
            shares = np.random.dirichlet(np.ones(len(values)), size=1)[0] * 100
            for val, share in zip(values, shares):
                dem_rows.append({
                    "year": year,
                    "region": region,
                    "demographic_category": cat,
                    "demographic_value": val,
                    "share_pct": round(share, 2),
                    # Optional: estimated count from share of total units sold
                    "estimated_units": int(round(units * share / 100))
                })

df_demo = pd.DataFrame(dem_rows)
df_demo = df_demo.sort_values(["region", "year", "demographic_category", "demographic_value"],
                              ascending=[True, False, True, True])

df_demo.to_csv("zx_cola_demographics.csv", index=False)

# ---------------------------
# Optional output confirmation
# ---------------------------
print("Sales/profit dataset saved to zx_cola_sales_profit.csv")
print("Demographics dataset saved to zx_cola_demographics.csv")
```
