```python
# dataset_template.py
import pandas as pd
from datetime import datetime

# ---------------------------------------------------------
# 1) Define columns for the dataset
# ---------------------------------------------------------
columns = [
    # Basic sale info
    "date",              # YYYY-MM-DD
    "country",           # UK or US
    "source",            # e.g., "eBay", "Retail", "Distributor"
    "sku",               # product code or name
    "pack_size",         # # cans per pack
    "units_sold",        # # packs sold
    "price_per_pack",    # in GBP or USD
    "revenue",           # computed or reported
    "cost_per_pack",     # in GBP or USD
    "cost_total",        # computed
    "profit_total",      # computed
    "profit_margin",     # computed, decimal or %
    # Demographics (optional, add only if available/allowed)
    "buyer_age_group",   # example: "18-24", "25-34", etc.
    "buyer_gender",      # e.g., "F", "M", "Other/NA"
    "city",              # e.g., "London", "New York"
    # Context flags
    "promo_flag",        # "Yes"/"No"
    "availability",      # "In stock", "Out of stock", "Limited"
    # Notes or free text
    "notes"             
]

# ---------------------------------------------------------
# 2) Create a sample dataframe with synthetic data
# ---------------------------------------------------------
data = [
    # A UK example row
    {
        "date": "2026-02-25",
        "country": "UK",
        "source": "eBay",
        "sku": "Vive Lemon Soda Zero Sugar 330ml",
        "pack_size": 24,
        "units_sold": 5,
        "price_per_pack": 15.98,   # GBP, from listing anchor
        "revenue": 5 * 15.98,      # compute total revenue
        "cost_per_pack": 10.50,    # hypothetical cost to seller
        "cost_total": 5 * 10.50,
        "profit_total": (5 * 15.98) - (5 * 10.50),
        "profit_margin": ((5 * 15.98) - (5 * 10.50)) / (5 * 15.98),
        "buyer_age_group": "25-34",
        "buyer_gender": "F",
        "city": "London",
        "promo_flag": "No",
        "availability": "In stock",
        "notes": "Regular listing; no coupon"
    },
    # A US example row - synthetic/placeholder
    {
        "date": "2026-02-26",
        "country": "US",
        "source": "Retail",
        "sku": "Vive Lemon Soda Zero Sugar 330ml",
        "pack_size": 12,
        "units_sold": 3,
        "price_per_pack": 12.99,    # USD hypothetical
        "revenue": 3 * 12.99,
        "cost_per_pack": 8.00,
        "cost_total": 3 * 8.00,
        "profit_total": (3 * 12.99) - (3 * 8.00),
        "profit_margin": ((3 * 12.99) - (3 * 8.00)) / (3 * 12.99),
        "buyer_age_group": "35-44",
        "buyer_gender": "M",
        "city": "New York",
        "promo_flag": "Yes",
        "availability": "Limited",
        "notes": "Intro promo bundle"
    },
    # Additional rows can be appended here...
]

df = pd.DataFrame(data, columns=columns)

# ---------------------------------------------------------
# 3) Sanity checks and basic calculations
# ---------------------------------------------------------
# Recompute derived fields for integrity
df["revenue"] = df["units_sold"] * df["price_per_pack"]
df["cost_total"] = df["units_sold"] * df["cost_per_pack"]
df["profit_total"] = df["revenue"] - df["cost_total"]
df["profit_margin"] = df["profit_total"] / df["revenue"]

# Optional: round numeric columns
numeric_cols = ["price_per_pack", "revenue", "cost_per_pack", 
                "cost_total", "profit_total", "profit_margin"]
df[numeric_cols] = df[numeric_cols].round(2)

# ---------------------------------------------------------
# 4) Export to CSV for UK and US separately
# ---------------------------------------------------------
today_str = datetime.today().strftime("%Y%m%d")

# Full dataset export
df.to_csv(f"vive_lemon_zero_sugar_sales_{today_str}.csv", index=False)

# Separate UK and US CSVs
df[df["country"] == "UK"].to_csv(f"vive_lemon_sales_UK_{today_str}.csv", index=False)
df[df["country"] == "US"].to_csv(f"vive_lemon_sales_US_{today_str}.csv", index=False)

print("CSV files saved.")
```

```python
def add_row(df, date, country, source, pack_size, units, price, cost,
            age_group=None, gender=None, city=None, promo=False, availability="In stock", notes=""):
    
    row = {
        "date": date,
        "country": country,
        "source": source,
        "sku": "Vive Lemon Soda Zero Sugar 330ml",
        "pack_size": pack_size,
        "units_sold": units,
        "price_per_pack": price,
        "revenue": units * price,
        "cost_per_pack": cost,
        "cost_total": units * cost,
        "profit_total": units * price - units * cost,
        "profit_margin": (units * price - units * cost) / (units * price) if units and price else 0,
        "buyer_age_group": age_group,
        "buyer_gender": gender,
        "city": city,
        "promo_flag": "Yes" if promo else "No",
        "availability": availability,
        "notes": notes
    }

    df = pd.concat([df, pd.DataFrame([row])], ignore_index=True)
    return df
```
